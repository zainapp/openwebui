## (3.0)/ INITIAL TRAINING

**WARNING**: Use this cell only when **starting from scratch**

```python
# Training from scratch!
#######################################################
# WARNING: this restarts from the INITIAL YOLO model!
#######################################################
!date

from ultralytics import YOLO

# Load the model from Ultralytics
model = YOLO('yolo12s.pt')

results = model.train(data='/content/drive/MyDrive/YOLO/datasets/medium/dataset.yaml', imgsz=1984, batch=1, epochs=5, plots=False, degrees=0.25, shear=0.1, fliplr=0.0, mosaic=0.0, rect=True)

# Save the 'detect' folder to Google drive
!cp -R --target-directory=/content/drive/MyDrive/YOLO/datasets/medium /content/runs/detect

!date
```

## 3/ Incremental training(s) from last model

**Input**:
- `last.pt` on drive

**Output**:
- `detect/train` folder (with `last.pt` and `last.onnx` models) on drive

***Important***  
Before launch, make sure that on Google drive:
- the `last.pt` file is the correct one
- the previous `detect` folder, if any, as been renamed to something different (for example `detect-N`)

**Performance**:
- Training yolo12s on our dataset (1600 images) takes about 14 minutes per epoch
- This means 4.5 hours for 20 epochs...

```python
# Starting time stamp
!date

from ultralytics import YOLO

# Load the last model
model = YOLO('/content/drive/MyDrive/YOLO/datasets/medium/last.pt')

# Train ...
results = model.train(batch=1, epochs=20, degrees=0.0, shear=0.0, fliplr=0.0, mosaic=0.0, rect=True)

# Export the model to ONNX format (next to the local input file last.pt)
model.export(format='onnx', imgsz=1984, dynamic=True, opset=12, simplify=True, device='cpu')

# Save the 'content' folder to Google drive
!cp -R --target-directory=/content/drive/MyDrive/YOLO/datasets/medium /content/runs/detect

# Ending time stamp
!date
```
